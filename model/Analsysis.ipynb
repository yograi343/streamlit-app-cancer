```python
from sklearn.model_selection import train_test_split
X=df.drop(['diagnosis'],axis=1)
y=df['diagnosis'].copy()
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,train_size=0.8,random_state=42)

len(X_train)/len(X)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled=scaler.fit_transform(X_train)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg= LogisticRegression(max_iter=1000)
```

```python
log_reg.fit(X_scaled,y_train)
```

```python
X_scaled_test = scaler.fit_transform(X_test)
y_predict = log_reg.predict(X_scaled_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score ,recall_score,f1_score,confusion_matrix,roc_auc_score,classification_report
```

```python
def displayscore(pred,label):
    accuracy = accuracy_score(pred,label)
    precision = precision_score(pred,label)
    recall = recall_score(pred,label)
    f1 = f1_score(pred,label)
    confusion =confusion_matrix(pred,label)
    roc_auc = roc_auc_score(pred,label)
    class_report = classification_report(pred,label)
    print('Accuracy:',accuracy)
    print('Precision:',precision)
    print('Recall:',recall)
    print('F1 Score:',f1)
    print('Classification Report:',class_report)
displayscore(y_predict,y_test)
```
